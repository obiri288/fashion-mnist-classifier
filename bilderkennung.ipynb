```python
# Schritt 1 & 2: Daten laden, untersuchen und vorverarbeiten

# Importiere TensorFlow und die Keras API
import tensorflow as tf
from tensorflow import keras

# Hilfsbibliotheken zum Plotten und für Mathematik
import numpy as np
import matplotlib.pyplot as plt

# Lade den Fashion MNIST Datensatz
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Definiere die Namen für die 10 Kleidungsklassen
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# --- Daten untersuchen ---
# Lass uns das erste Bild im Trainingsdatensatz ansehen
plt.figure()
plt.imshow(train_images[0], cmap=plt.cm.binary)
plt.colorbar()
plt.grid(False)
plt.title(f"Label: {class_names[train_labels[0]]}")
plt.show()

# --- Daten vorverarbeiten ---
# Normalisiere die Pixelwerte von 0-255 auf 0-1
train_images = train_images / 255.0
test_images = test_images / 255.0

# --- Überprüfen ---
# Zeigen wir die ersten 25 Bilder nach der Normalisierung an,
# um zu überprüfen, ob alles geklappt hat.
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()

# --- Schritt 3: Das Modell bauen ---

model = keras.Sequential([
    # Wandelt die 2D-Bilder (28x28) in einen 1D-Vektor (784) um
    keras.layers.Flatten(input_shape=(28, 28)),

    # Eine Schicht mit 128 Neuronen, die Muster lernt
    keras.layers.Dense(128, activation='relu'),

    # Die Ausgabeschicht mit 10 Neuronen (eins pro Klasse), die Wahrscheinlichkeiten ausgibt
    keras.layers.Dense(10, activation='softmax')
])

print("Modell-Architektur erfolgreich erstellt!")

# --- Schritt 4: Das Modell kompilieren ---

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Zeige eine Zusammenfassung des Modells an
model.summary()

# --- Schritt 4: Das Modell kompilieren ---

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Zeige eine Zusammenfassung des Modells an
model.summary()

# --- Schritt 5: Das Modell trainieren ---

# Füttere das Modell mit den Trainingsdaten
# epochs=10 bedeutet, dass das Modell den gesamten Datensatz 10 Mal "sieht"
history = model.fit(train_images, train_labels, epochs=10)

print("\nTraining abgeschlossen!")

# --- Schritt 6: Genauigkeit auf dem Test-Set bewerten ---

test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print('\nGenauigkeit auf dem Test-Set:', test_acc)


# --- Vorhersagen für das Test-Set treffen ---
predictions = model.predict(test_images)


# --- Eine einzelne Vorhersage visualisieren ---
# Wir definieren zwei kleine Hilfsfunktionen, um die Plots schöner zu machen
def plot_image(i, predictions_array, true_label, img):
    true_label, img = true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel(f"{class_names[predicted_label]} {100*np.max(predictions_array):2.0f}% ({class_names[true_label]})",
               color=color)

def plot_value_array(i, predictions_array, true_label):
    true_label = true_label[i]
    plt.grid(False)
    plt.xticks(range(10))
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions_array, color="#777777")
    plt.ylim([0, 1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')

# Wähle ein Bild zum Anzeigen aus (z.B. das 13. Bild)
i = 15
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions[i], test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions[i], test_labels)
plt.show()
```
